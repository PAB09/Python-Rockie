Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (40 pts)

MAking age group:

```python
child = Titanic.loc[Titanic['Age'] < 14]
teen = Titanic.loc[Titanic['Age'] >= 14]
teen = teen.loc[teen['Age'] < 21]
adult = Titanic.loc[Titanic['Age'] >= 21]
adult = adult.loc[adult['Age'] < 65]
senior = Titanic.loc[Titanic['Age'] >= 65]
```

After making age group we add one column in dataset 

```python
Titanic["Age_range"] = ""

# loop through to fill out new column properly
for x in range(1, len(Titanic)+1):
    if (Titanic.Age[x] < 14.0):
        Titanic.Age_range[x] = "child"
    elif (Titanic.Age[x] >= 14.0) and (Titanic.Age[x] < 21.0):
        Titanic.Age_range[x] = "adolescent"
    elif (Titanic.Age[x] >= 21.0) and (Titanic.Age[x] < 65):
        Titanic.Age_range[x] = "adult"
    elif (Titanic.Age[x] >= 65.0):
        Titanic.Age_range[x] = "senior"
    else:
        Titanic.Age_range[x] = "NaN"
```
