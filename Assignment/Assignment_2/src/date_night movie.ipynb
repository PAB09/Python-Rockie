```python
import pandas as pd
import os
```

BY use of get_movie_data() we retrive data from dat files.

```python
def get_movie_data():
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_table(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames)
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_table(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames)
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_table(os.path.join('../data', 'movies.dat'), 
                           sep='::', header=None, names=mnames)

    return users, ratings, movies
```

```python
users, ratings, movies = get_movie_data()
```

Now , we seperate year from movie name

```python
tmp = movies.title.str.extract('(.*) \(([0-9]+)\)')
tmp.apply(lambda x:x[0] if len(x) > 0 else None)
tmp.apply(lambda x: x[0][:40] if len(x) > 0 else None)
```

```python
movies['year'] = tmp[1]
movies['short_title'] = tmp[0]
```

# Now Merging 3 datasets

Merge users and ratings dataset into mUR

```python
mUR=pd.merge(users , ratings, on='user_id')
mUR
```

Merge mUR and movies dataset into mURM

```python
mURM=pd.merge(mUR,movies, on='movie_id')
mURM
```

By using groupby function and size method, we summarize movie title and it's total rates.

```python
RT1= mURM.groupby('short_title',sort=False,as_index=False).size()
RT1.head()
```

```python
RT2= mURM.groupby('short_title',sort=False,as_index=False)['rating'].mean()## with mean() we calculate exact rating of movie.
RT2.head()
```

```python
RT2['size']=RT1['size']## combine total ratings and exact ratings with each other.
RT2
```

```python
HR=RT2[RT2['size']>1000]## maming dataset(HR) that includes moive name whose size are above 10000. 
HR
```

```python
HR=HR.sort_values(by=['size'],ascending=False)##sorting values for movie who is having highest number of rating size.
HR
```
